```python
# ---------------------------------    DATA TYPE            ------------------------------

# A data type is a classification that defines the kind of value a variable or expression can hold. 
# It tells the computer how to interpret and work with that data. 

# Integer (int): Represents whole numbers, like 5, -10, or 100.

# Float (float): Represents numbers with decimal points, like 3.14, -0.5, or 2.0.

# String (str):  A value enclosed betweem single quotes ( ' ' ) or double quotes ( " " )
#Represents text or sequences of characters , like "hello", "Python", or "123".

# Boolean (bool): Represents either True or False, and is often used for making decisions in code.

# List: Represents an ordered collection of items, like [1, 2, 3] or ['apple', 'banana', 'cherry'].

# Tuple: Similar to a list, but immutable (cannot be changed after creation), like (1, 2, 3) or ('red', 'green', 'blue').

# Dictionary: Represents a collection of key-value pairs, like {'name': 'Alice', 'age': 25}.

# Set: Represents an unordered collection of unique items, like {1, 2, 3} or {'apple', 'banana'}.

# NoneType (None): Represents the absence of a value, often used to indicate 
# that something doesn't exist or hasn't been assigned yet.

name = "Bob" # STR
age = 15 # INT
height = 12.3 # FLOAT
studying = True # BOOLEAN
test ="12.3"

print("name =",name,'\n',"age =",age,'\n',"height =",height,'\n',"studying =",studying,'\n',test)
```

```python
#---------------------         SUBSCRIPT []     -------------------------\
# pulling out a particular character from string is called subscript .
# number between the square bracket [] determines which character you want to pull out .
# first number will always start with zero (0) , because programing language work with binary language and 
# its work with zero(0) and (1) .
# -------
# A Subscript refers to using square brackets [] to access a specific element or 
# elements within a collection, like a list, tuple, string, or dictionary.
# # It's a way to indicate which item you want from the collection, based on its position or key.

print("hello"[4])

# --------Here's how subscripts work with different types of collections:

#\\\\\\\\   Lists and Tuples:

# Lists and tuples are ordered collections where each item has an index (position) starting from 0.
# You use a subscript to access an item by its index.

my_list = [10, 20, 30]
second_item = my_list[1]  # Gets the second item (20)
print(second_item)

#\\\\\\\ Strings:

# A string is a sequence of characters, and each character has an index.
# You use subscripts to access characters by their index.

my_string = "hello"
first_char = my_string[0]  # Gets the first character ("h")
print(first_char)

#\\\\\\\\\\ Dictionaries:

# A dictionary is a collection of key-value pairs, and you use the keys (not indices) as subscripts to access values.

my_dict = {'name': 'Alice', 'age': 25}
person_name = my_dict['name']  # Gets the value associated with the key 'name' ("Alice")
print(person_name)

# \\\\ NOTE ----
# subscripts typically start counting from 0 for lists, tuples, and strings. 
# Negative subscripts count from the end , -1 refers to the last item. counting start from -1 for the last item .
# For dictionaries, subscripts correspond to the keys . 
```

```python
# -------------------      TYPE CONVERSION OR TYPE CASTING     ---------------------
# Type conversion, also known as type casting, in Python refers to the process of changing the data type of a value
# from one type to another. This can be useful when you need to perform operations or comparisons that involve different data types.
# Python provides several built-in functions to perform type conversion.

# Here are some common type conversion functions:

# \\\\\\\\\   int(): Converts a value to an integer data type.

float_number = 5.12
integer = int(float_number) # Converts the float to an integer (5)
print(integer) 

#\\\\\\\\\ float(): Converts a value to a floating-point (decimal) data type .

integer = 7
float_num = float(integer) # converts the integer to a float number (7.0)
print(float_num)

#\\\\\\\\\ str(): Converts a value to a string data type .

int = 77 
string = str(int) # converts the integer to a string . 
print(string)

# \\\\\\\\\  list(), tuple(), set(): Converts a sequence (like a string or another iterable) to a list, tuple, or set data type.

text = "hello"
list_of_chars = list(text)  # Converts the string to a list (['h', 'e', 'l', 'l', 'o'])
print(list_of_chars)

#\\\\\\\\ bool(): Converts a value to a boolean data type.

value = 10 
boolean = bool(value) # Converts the value to a boolean (right)
print(boolean)

# Remember that not all types can be converted to all other types. For example,
# converting a string that contains letters to an integer wouldn't work directly.
# It's important to use type conversion appropriately and be aware of potential data loss or errors that can occur during conversion.

user_input = input("Enter a number: ")
number = int(user_input)  # Converts the user input (string) to an integer
if number > 10:
    print("The number is greater than 10.")
else:
    print("The number is not greater than 10.") 



```

```python
# ---------------------------         MATHMETICAL OPERATIONS         -----------------------------------


# Operators are special symbols in Python that carry out arithmetic or logical computation. 
# The value that the operator operates on is called the operand. operands are the values or variables that are used in operations, 
# such as arithmetic or logical operations. 

# **Arithmetic operators**

# Arithmetic operators are used to perform mathematical operations like addition, subtraction, multiplication, etc.

#\\\\\\\\\ Addition (+): Adds two numbers together.

# sum_result = 5 + 3  # 8

# \\\\\\\\ Subtraction (-): Subtracts the second number from the first.

# difference = 10 - 4  # 6

# \\\\\\\\\\ Multiplication (*): Multiplies two numbers.

# product = 7 * 2  # 14

# \\\\\\\\\\ Division (/): Divides the first number by the second. Always returns a float.

# quotient = 15 / 3  # 5.0

#------ Whenever you divide you will always end up with floating number .

# \\\\\\\\  Integer Division or floor division (//):
#  Divides the first number by the second and returns an integer (rounded down). 
# The // operator performs division and then rounds down the result to the nearest integer.
# if you want your output as int . use floor division .

# quotient = 17 // 5
# print(quotient)  # Output: 3
# In this example, 17 is divided by 5, resulting in 3.4, but the // operator performs floor division and rounds down to 3.


# \\\\\\\\\Modulus (%): Returns the remainder of division between two numbers.

# remainder = 17 % 5  # 2

# \\\\\\\\\\Exponentiation (**): Raises the first number to the power of the second.

# power = 2 ** 3  # 8
# powerr = 3 ** 2 # 9

#\\\\\\\\\\  NOTE-
# in mathematical operations if you have more than one operation on the same line of code , then there's a certain level of priority.

#  PEMDASLR - Order of execution
#  p = parenthesis ()
#  E = Exponentiation **
#  M = Multipicati *
#  D = Divide /
#  A = Add + 
#  S = Subtract -
#  LR = The calculation goes from left to right
```

```python
# -----------------       ROUND() FUNCTION      -------------------
# round functionis a built-in function that allows you to round a number to a specified number of decimal places. 
# It's commonly used when you want to simplify or limit the precision of a floating-point number.

# \\\\  rounded_number = round(number, num_decimal_places)

# Where:

# number is the floating-point number you want to round.
# num_decimal_places is the number of decimal places to round the number to.
# rounded_number is the result after rounding

number = 8.45676776
rounded_number = round(number,2)
print(rounded_number)

value=5.1234
rounded_value=print(round(value))
print(type(rounded_value))


#\\\\\\\\\\\\\\\\\  NOTE --
# Keep in mind that the round() function returns a floating-point number, 
# even if you're rounding to a whole number.
#  If you want an integer result, you can explicitly convert the rounded number to an integer using int().
```

```python
# # -----------------   ASSIGNMENT OPERATORS ()   --------------------------------
# # Assignment operators in Python are used to assign values to variables. 
# # They allow you to update the value of a variable by performing an operation on its current value and a new value.
# # Assignment operators combine the operation with the assignment in a concise way. 

# # Here are some common assignment operators in Python:

#\\\\\\\\\\  
# '=' (Assignment Operator): 
# Assigns the value on the right to the variable on the left.

# x = 5

#\\\\\\\\\\  
# '+=' (Addition Assignment Operator): 
# Adds the value on the right to the current value of the variable and assigns the result to the variable.

# x += 3  # Equivalent to x = x + 3

# \\\\\\\\\ 
# '-=' (Subtraction Assignment Operator): 
# Subtracts the value on the right from the current value of the variable and assigns the result to the variable.

# y -= 2  # Equivalent to y = y - 2

#\\\\\\\\\ 
# '*=' (Multiplication Assignment Operator):
#  Multiplies the current value of the variable by the value on the right and assigns the result to the variable.

# z *= 4  # Equivalent to z = z * 4

# \\\\\\\\\\ 
# '/= '(Division Assignment Operator):
#  Divides the current value of the variable by the value on the right and assigns the result to the variable.

# w /= 2  # Equivalent to w = w / 2

# \\\\\\\\\\
# '%=' (Modulus Assignment Operator): 
# Calculates the remainder of dividing the current value of the variable by the value on the right and assigns the result to the variable.

# a %= 7  # Equivalent to a = a % 7

#\\\\\\\\\\\\\\
# '//=' (Floor Division Assignment Operator):
#  Performs floor division on the current value of the variable and the value on the right, then assigns the result to the variable.

# b //= 3  # Equivalent to b = b // 3

# \\\\\\\\\\\\\\\
# '**=' (Exponentiation Assignment Operator):
#  Raises the current value of the variable to the power of the value on the right and assigns the result to the variable.

# c **= 2  # Equivalent to c = c ** 2

```

```python
# ---------------------  F STRING   ----------------------------------

# F string makes it really easy to mix strings and differnet data types .
# An F-string, short for "formatted string literal," is a way to create strings in Python that allows you to embed expressions inside them. 
# It's a convenient and readable way to build strings that include variables, calculations, or other Python expressions.

# --- how you do it-- 

# Start your string with the letter f or F right before the opening quote.
# Inside the string, you can use curly braces {} to enclose expressions you want to include.
# The expressions inside the curly braces will be evaluated and replaced with their values when the string is created.

name = "shobhit"
age = 25 
score = 250
is_winning = True 
print(f"his name is {name}, he is {age} . he loves to play cricket , his high score is {score}, he is winning is {is_winning}")

x = 20
y = 30
result = f"The sum of {x} and {y} is {x+y}"
print(result)
```
