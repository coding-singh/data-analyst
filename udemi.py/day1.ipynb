```python
#       -------------------------    INPUT FUNCTION    ------------------------------------------------
#  input function allows  a program to receive a input from a user .  This function prompts the user to input the value . 
# the captured input can then be used by the program for various purposes, such as calculations, processing, or displaying information.
#  input ( " a prompt for  the user ") - what kind of data we want .
# ""---- in input function we gave user a prompt nd according to the prompt user input some value (data) .
# nd in output data will be according to the prompt inputed by user. ""

name = input ( "what is your name")
print(name)

# output will be the name as entered by the user .

#         ----------     Convert input strings to integers   ----------------
# In python, the input() function returns a string, regardless of whether you enter a number or any other input.
# When you use the + operator on two strings, it concatenates them together to create a new string,
# not perform addition as you might expect for numbers.
# To get the sum of two numbers, you need to convert the input strings to integers - int() or floats - float() before performing addition.
# You can do this using the int() function for integer conversion or float() function for float conversion.
# \\\\ NOTE - Input function always returns string data type .

a = input("enter a ")
b = input("enter b ")


a = int(a)
b = int(b)

c = a + b
print("sum is", c)
```

```python
# ----------------------------   LEN () FUNCTION   ----------------------------------------------
# The len() function in Python is a built-in function that helps you find out the length of certain objects, such as strings, lists, or tuples.
#  It tells you how many items or characters are present in the object.
# length = len(some_object)
# WHERE --
# -- some_object is the object you want to measure the length of.
# --length is the variable that will hold the result, which is the number of items or characters in some_object
text = "Hello, world!"
length = len(text)
print("The length of the text is:", length)

# len() function doesn't like working with integers .

```

```python
#  ---------------------------    MAX () FUNCTION   --------------------------------------------------------
# The max() function in Python is used to find the largest value among a group of numbers or elements. 
# It helps you quickly figure out the biggest item in a list, a set of numbers, or any other sequence.
a=(10,12,30,40,50)
print(max(a)) 

fruits = ("banana","apple","mango","cherry" ,"guava")
print(max(fruits))
```

```python
# --------------------      MIN () FUNCTION     ---------------------------------
# The min() function in Python is used to find the smallest value among a group of numbers or elements.
# It helps you quickly determine the minimum item in a list, a set of numbers, or any other sequence. 

a=(10,50,100,30,65)
print(min(a))

city= ("delhi ", "bhopal","gwalior" , "faridabad" , "orai")
min_city= min(city)
print(min_city)

####  CASE - 1 - If we have a list of string and integer and we want to use max function then itt will throw error that string and integer 
####  value comparison is not possible .

#### CASE - 2 - If string value comparison is there .in such case it will be compared as A to Z in ascending order .


```
