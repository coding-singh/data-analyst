```python
# -----------------------      NESTED ELIF STATEMENT   -----------------------------------

# Nested if and elif statements in Python are used when you want to check multiple conditions in a hierarchical or sequential manner.
# They allow you to create more complex decision-making structures by testing one condition inside another.
# A nested if-elif statement in Python is a way to include one or more if-elif statements inside another.

# ///////  NOTE -
# you can add as many 'elif' condidtion  in between if and else  as much you want .

# if condition1:
#     # Code to execute if condition1 is True
#     if condition2:
#         # Code to execute if condition2 is also True
#     elif condition3:
#         # Code to execute if condition3 is True (inside the inner if)
#     else:
#         # Code to execute if none of the conditions are True (inside the inner if)
# else:
#     # Code to execute if condition1 is False .

print (" welcome to rollercoaster ride ")
height = int(input(" tell me your height in cm "))
if height >= 120:
    print("you can ride on rollercoaster")
    age = int(input("tell me your age"))
    if age < 12 :
        print( " please pay $ 5 " )
    elif age <= 18:
        print("please pay $ 7")
    else :
        print(" please pay $ 12 ")
else :
    print("better luck next time")
```

```python
# ////////   LEAP YEAR PROGRAM --------------

# Write a program that works out whether if a given year is a leap year.
# A normal year has 365 days, leap years have 366, with an extra day in February.
# The reason why we have leap years is really fascinating, this video does it more justice:

# This is how you work out whether if a particular year is a leap year.

# on every year that is evenly divisible by 4 
# **except** every year that is evenly divisible by 100 
# **unless** the year is also evenly divisible by 400

year = int(input(" which year you want to know"))
if(year%4)==0:
    if(year%100)==0:
        if(year%400)==0:
            print("its a leap year")
        else:
            print("its not a leap year")
    else:
        print("its a leap year") 
else:
    print("its not a leap year")

# if year % 4 = 0 and year % 100 != 0 then its a leap year , 
# if year % 400 = 0 its a leap year
```
